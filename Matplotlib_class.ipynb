```python
fig = plt.figure()
ax = plt.axes(projection='3d')

z = np.linspace(0,1,100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)


ax.plot3D(x, y, z, 'blue')
plt.title('3D line plot')

plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

z = np.linspace(0,1,100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
c = x + y

ax.scatter(x, y, z, c=c)
plt.title('3D line plot')

plt.show()
```

```python
x = np.outer(np.linspace(-2,2,30), np.ones(30))
y = x.copy().T # here T is for transformation
z = np.cos(x ** 2 + y ** 2)

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x, y, z, cmap='viridis')
plt.title('3D surface plot')

plt.show()
```

# Section 6: Widgets

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import widgets
from IPython.display import display
import os
```

```python
def exponent(maxIteration):
    x = np.arange(5 * maxIteration)
    y = np.exp(x)
    
    plt.scatter(x, y, color='blue', marker='o')
    plt.title('Widgets')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()
    
maxIteration = widgets.Play(min=1, max=1000, value=1, description='Iter')

widgets.interactive(exponent, maxIteration=maxIteration)
```
